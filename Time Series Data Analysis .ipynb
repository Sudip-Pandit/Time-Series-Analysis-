### Resampling 
Resampling is the process of taking the mean/average 
by changing its interval.

```python
### Resampling of data month wise
```

```python
df['Open'].resample('M').mean()
```

```python
df['Open'].resample('M').mean().plot(figsize=(8,7))
```

```python
df[['Open', 'Close']].resample('M').mean().plot(figsize=(8,7))
```

```python
df[['Open', 'Close']].resample('M').mean().plot.bar(figsize=(8,7))
```

```python
#Bar horizontal 
```

```python
df['Open'].resample('M').mean().plot.barh(figsize=(8,7))
```

### Weekely Resample

```python
df['Open'].resample('W').mean().head()
```

```python
df['Open'].resample('W').mean().plot(figsize=(8,7))
```

```python
df['Open'].resample('W').mean().head().plot.barh(figsize=(8,7))
```

```python
df[['Open', 'Close']].resample('W').mean().head().plot.barh(figsize=(8,7))
```

### 15 Days Period - SM 

```python
df[['Open', 'Close']].resample('SM').mean().head().plot.barh(figsize=(8,7))
```

```python
df['Open'].resample('SM').mean().head()
```

```python
df['Open'].resample('SM').mean().plot(figsize=(8,7))
```
