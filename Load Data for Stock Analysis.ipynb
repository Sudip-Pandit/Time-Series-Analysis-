```python
start = datetime.datetime(2019,1,1)
end = datetime.datetime(2019,4,17)
```

```python
google = web.DataReader("GOOGL", 'yahoo', start, end)
```

```python
google.head(3)
```

```python
google['Open'].head()
```

```python
google['Open'].plot()
```

```python
#Let's compare open, close and high prices
import matplotlib.pyplot as plt
google['Open'].plot(label = 'GOOGL Open price')
google['Close'].plot(label = 'GOOGL Close price')
google['High'].plot(label = 'GOOGL High price')
plt.show()
```

```python
#Let's label it
google['Open'].plot(label = 'GOOGL Open price')
google['Close'].plot(label = 'GOOGL Close price')
google['High'].plot(label = 'GOOGL High price')
plt.legend()
plt.title('Google stock price')
plt.ylabel('Overall stock price')
plt.xlabel('Date')
plt.show()
```
